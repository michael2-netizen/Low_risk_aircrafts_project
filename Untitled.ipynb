## Low Risk Aircraft Project ##

### Project Goal ###

The project aims to determine the potentially low risk aircraft an investor may consider when interested in purchasing aircraft. To achieve this, the project aims to translate findings into actionable insights that would easily enable an investor make a good decision.

### Data source ###

This data has been derived  from the National Transportation Safety Board that includes aviation accident data from 1962 to 2023 about civil aviation accidents and selected incidents in the United States and international waters. For more information on the data [click this link](https://www.kaggle.com/datasets/khsamaha/aviation-accident-database-synopses)

### Data Visualization ###

```python
# Convert Event.Date to datetime and extract the year
df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')
df['Year'] = df['Event.Date'].dt.year
```

```python
# Calculate frequencies
accident_counts = df['Year'].value_counts().sort_index()
```

```python
# Plot the graph
plt.figure(figsize=(12, 6))
plt.bar(accident_counts.index, accident_counts.values, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.title('Frequency of Aviation Accidents per Year')
plt.show()
```

```python
 # occurrences of each Investigation.Type
accident_types = df['Investigation.Type'].value_counts()
```

```python
 # Create a pie chart
plt.figure(figsize=(10, 7))
plt.pie(accident_types, labels=accident_types.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribution of Accident Types')
plt.axis('equal')  
plt.show()
```

```python
# Create a line graph
plt.figure(figsize=(12, 6))
plt.plot(accident_counts.index, accident_counts.values, marker='o', color='blue', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.title('Trends in Aviation Accidents Over Time')
plt.show()
```

```python
# Assuming Total.Fatal.Injuries and Total.Serious.Injuries are our columns of interest
df['Total.Injuries'] = df['Total.Fatal.Injuries'].fillna(0) + df['Total.Serious.Injuries'].fillna(0)
```

```python
# Creating a scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(df['Year'], df['Total.Injuries'], alpha=0.6, color='orange')
plt.title('Total Injuries in Aviation Accidents Over Time')
plt.xlabel('Year')
plt.ylabel('Total Injuries')
plt.show()
```

```python
# Count the number of aircraft per number of engines
engine_data = df['Number.of.Engines'].value_counts().sort_index()
engine_data
```

```python
#Create a scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(engine_data.index, engine_data.values, alpha=0.6, color='orange')
plt.title('Number of Engines vs. Aircraft Count')
plt.xlabel('Number of Engines')
plt.ylabel('Number of Aircraft')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
#Count the number of accidents per number of engines
engine_accidents = df['Number.of.Engines'].value_counts().sort_index()
engine_accidents
```

```python
#Create a bar chart
plt.figure(figsize=(12, 6))
engine_accidents.plot(kind='bar', color='lightcoral')
plt.title('Total Accidents by Number of Engines')
plt.xlabel('Number of Engines')
plt.ylabel('Total Number of Accidents')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#Count occurrences of each Engine Type
engine_counts = df['Engine.Type'].value_counts()
engine_counts
```

```python
#Create a bar chart
plt.figure(figsize=(12, 6))
engine_counts.plot(kind='bar', color='lightgreen')
plt.title('Number of Aircraft by Engine Type')
plt.xlabel('Engine Type')
plt.ylabel('Number of Aircraft')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#Create a bar chart
plt.figure(figsize=(12, 6))
engine_counts.plot(kind='bar', color='lightgreen')
plt.title('Purpose of flight by Engine Type')
plt.xlabel('Engine Type')
plt.ylabel('Purpose.of.flight')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
